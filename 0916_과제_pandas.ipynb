### DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환

DataFrame과 넘파이 ndarray 상호 간의 변환은 매우 빈번하게 발생

> #### 넘파이 ndarray, 리스트, 딕셔너리를 DataFrame으로 변환하기

DataFrame은 칼럼명을 가지고 있어 편하게 데이터 핸들링 가능
-> 데이터와 칼럼명 리스트를 입력 받아 쉽게 생성

##### 1차원 형태의 리스트와 넘파이 ndarray  => DataFrame

````python
import numpy as np
col_name1=['col1']    # 1차원이므로 칼럼 한개만 필요
list1 = [1,2,3]
array1 = np.array(list1)
print('array1 shape:',array1.shape)

# 리스트를 이용해 DataFrame 생성
df_list1 = pd.DataFrame(list1,columns=col_name1)
print('1차원 리스트로 만든 DataFrame:\n',df_list1)

# 넘파이 ndarray를 이용해 DataFrame 생성
df_array1 = pd.DataFrame(array1,columns=col_name1)
print('1차원 ndarray로 만든 DataFrame:\n',df_array1)
````

##### 2차원 형태의 리스트와 넘파이 ndarray (2행 3열 형태) => DataFrame

````python
# 3개의 칼럼명이 필요함
col_name2 = ['col1','col2','col3']

# 2행x3열 형태의 리스트와 ndarray 생성한 뒤 이를 DataFrame으로 변환
list2 = [[1,2,3],
         [11,12,13]]
array2 = np.array(list2)
print('array2 shape:',array2.shape)

# 리스트를 이용해 DataFrame 생성
df_list2 = pd.DataFrame(list2,columns=col_name2)
print('2차원 리스트로 만든 DataFrame:\n',df_list2)

# 넘파이 ndarray를 이용해 DataFrame 생성
df_array2 = pd.DataFrame(array2,columns=col_name2)
print('2차원 ndarray로 만든 DataFrame:\n',df_array2)
````

##### 딕셔너리   => DataFrame

키(Key) 값은 칼럼명, 값(Value)은 각 칼럼 데이터로 변환

````python
# Key는 문자열 칼럼명으로 매핑, Value는 리스트 형(또는 ndarray) 칼럼 데이터로 매핑
dict = {'col1':[1,11],'col2':[2,22],'col3':[3,33]}
df_dict = pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame:\n',df_dict)
````

>#### DataFrame을 넘파이 ndarray, 리스트, 딕셔너리로 변환하기

많은 머신러닝 패키지가 넘파이 ndarray 사용 -> 넘파이 ndarray로 변환하는 경우 빈번하게 발생

##### DataFrame => 넘파이 ndarray

* DataFrame 객체의 **values** 이용

````python
# DataFrame을 ndarray로 변환
array3 = df_dict.values
print('df_dict.values 타입:',type(array3),'df_dict_values shape:',array3.shape)
print(array3)
````

##### DataFrame => 리스트와 딕셔너리

* 리스트로의 변환 = **values**로 얻은 ndarray에 **tolist()** 호출
* 딕셔너리로의 변환 = DataFrame 객체의 **to_dict()** 메서드 호출, 인자로 'list' 입력

````python
# DataFrame을 리스트로 변환
list3 = df_dict.values.tolist()
print('df_dict.values.tolist() 타입:',type(list3))
print(list3)

# DataFrame을 딕셔너리로 변환
dict3 = df_dict.to_dict('list')
print('\n df_dict.to_dict() 타입:',type(dict3))
print(dict3)
````

### DataFrame 데이터 삭제


<정리>
>* DataFrame의 [] 연산자는 넘파이의 []나 Series의 []와 다르다
>* [] 내 입력 값은 칼럼명을 지정해 칼럼 지정 연산에 사용하거나 불린 인덱스 용도로만 사용
>* 슬라이싱 연산으로 데이터 추출하는 방법은 사용x

> #### DataFrame ix[] 연산자

**ix[]** : 행 위치, 열 위치 지정으로 원하는 위치의 데이터 추출

행 위치 지정 = 인덱스 값 입력

열 위치 지정 = 칼럼 명 지정(칼럼 명칭 기반 인덱싱) / 칼럼의 위치 값 지정(칼럼 위치 기반 인덱싱)

=> ix[] 는 현재 판다스에서 사라져(사용시 error) => loc[]와 iloc[] 새로 만들어져

>#### 명칭 기반 인덱싱과 위치 기반 인덱싱의 구분

DataFrame의 인덱스값 = 명칭 기반 인덱싱

ix[]의 경우 행과 열 위치에 명칭과 위치 기반 인덱싱 모두 허용

But 행 위치에 적용되는 인덱스값과 위치 기반 인덱싱이 integer형일 때 혼선 초래 우려 

=> 명칭 기반 인덱싱 연산자 loc[]와 위치 기반 인덱싱 연산자 iloc[] 구분

>#### DataFrame iloc[] 연산자

**iloc[]** : 위치 기반 인덱싱 연산자
-> 행과 열 값으로 <u>integer 또는 integer형의 슬라이싱, 팬시 리스트 값</u>을 입력해줘야해

+) 불린 인덱싱은 제공x

````python
data = {'Name':['Chulmin','Eunkyung','Jinwoong','Soobeom'],
        'Year':[2011,2016,2015,2015],
        'Gender':['Male','Female','Male','Male']
        }
data_df = pd.DataFrame(data,index=['one','two','three','four'])
data_df
````

### apply lambda 식으로 데이터 가공

칼럼에 복잡한 데이터 가공이 필요한 경우 **apply lambda** 이용

* lambda : 파이썬에서 함수형 프로그래밍 지원하기 위해 만들어짐

   -> 함수의 선언과 함수 내의 처리를 한 줄의 식으로 쉽게 변환하는 식

````python
lambda_square = lambda x : x**2         # ':' 기준으로 왼쪽이 입력 인자, 오른쪽이 입력 인자의 계산식이자 반환 값
print('3의 제곱은:',lambda_square(3))
````

* 여러 개의 값을 입력 인자로 사용해야 할 경우 **map()** 함수 결합해 사용

````python
a = [1,2,3]
squares = map(lambda x : x**2, a)
list(squares)
````

````python
# DataFrame의 apply에 lambda 식 적용
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name','Name_len']].head(3)
````

````python
# if else 절 사용
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else 'Adult')
titanic_df[['Age','Child_Adult']].head(8)
````

* if else 사용할 때, if식보다 반환 값을 먼저 기술(':'기호의 오른편에 반환 값 있어야해) / else의 경우 else 이후에 반환 값

* else if는 지원하지 않는다 -> else절을 ()로 내포해 ()내에서 다시 if else 적용해 사용

````python
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x<15 else ('Adult' if x<=60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
````

````python
# else를 계속 내포해 쓰기 부담 -> 별도의 함수 만드는 게 나아
# 나이에 따라 세분화된 분류를 수행하는 함수 생성
def get_category(age):
  cat=''
  if age<=5 : cat='Baby'
  elif age<=12 : cat='Child'
  elif age<=18 : cat='Teenager'
  elif age<=25 : cat='Student'
  elif age<=35 : cat='Young Adult'
  elif age<=60 : cat='Adult'
  else : cat='Elderly'
  return cat

#lambda 식에 위에서 생성한 get_category()함수를 반환값으로 지정
# get_category(x)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age','Age_cat']].head()
````
